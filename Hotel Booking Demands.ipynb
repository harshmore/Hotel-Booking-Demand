# Input Split

```python
X = data.drop(columns = ["arrival_date_year", "arrival_date_week_number", "is_canceled", 'company', "arrival_date_month", 'reserved_room_type',"customer_type",'reservation_status_date'])
y = data["is_canceled"]
```

# Model Training

```python
from sklearn.model_selection import  train_test_split,cross_val_score

def train(model, X, y) :
    # train test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify = y)


    # model fitting
    model[1].fit(X_train, y_train)

    # cross validation
    cv_score = cross_val_score(model[1], X, y, cv=5)
    
    print(model[0])
    print("Accuracy:", model[1].score(X_test,y_test))
    print("CV Score:", np.mean(cv_score))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

models = [('LogisticRegression',LogisticRegression()),
         ('KNN',KNeighborsClassifier(n_neighbors=5)),
         ('DecisionTree',DecisionTreeClassifier()),
         ('RandomForest',RandomForestClassifier())]

for model in models:
    train(model,X,y)
    print()
```
